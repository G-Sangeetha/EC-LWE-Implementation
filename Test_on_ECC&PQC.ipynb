```python
import time
import os
import pandas as pd
from cryptography.hazmat.primitives.asymmetric import ec
from cryptography.hazmat.primitives import hashes
from cryptography.hazmat.primitives.kdf.pbkdf2 import PBKDF2HMAC
from cryptography.hazmat.primitives.ciphers import Cipher, algorithms, modes
from cryptography.hazmat.primitives import padding

# Define message sizes in bytes
message_sizes = [16, 32, 64, 128, 256, 512, 1024, 2048, 4096, 8192, 16384, 32768, 65536, 131072, 262144]

# File to store results
filename = "/content/performance_results.csv"

# Load existing data if available
if os.path.exists(filename):
    df = pd.read_csv(filename)
else:
    df = pd.DataFrame(columns=["Scheme", "Message Size", "Key Generation Time",
                               "Encryption Time", "Decryption Time", "Ciphertext Size"])

# Filter existing ECC results to avoid duplication
existing_ecc_sizes = set(df[df["Scheme"] == "ECC"]["Message Size"]) if "Scheme" in df.columns else set()

for size in message_sizes:
    if size in existing_ecc_sizes:
        print(f"‚è≠ Skipping ECC test for message size {size}, already in results.")
        continue  # Avoid duplicating ECC results

    # Generate random message with alphanumeric and symbols
    message = os.urandom(size)  # Random bytes as message

    # Key Generation for Sender
    start_time = time.time()
    private_key_sender = ec.generate_private_key(ec.SECP256R1())
    public_key_sender = private_key_sender.public_key()
    key_gen_time_sender = time.time() - start_time

    # Key Generation for Receiver
    start_time = time.time()
    private_key_receiver = ec.generate_private_key(ec.SECP256R1())
    public_key_receiver = private_key_receiver.public_key()
    key_gen_time_receiver = time.time() - start_time

    # ECDH Key Exchange to derive shared secret
    start_time = time.time()
    shared_secret_sender = private_key_sender.exchange(ec.ECDH(), public_key_receiver)
    shared_secret_receiver = private_key_receiver.exchange(ec.ECDH(), public_key_sender)
    key_exchange_time = time.time() - start_time

    assert shared_secret_sender == shared_secret_receiver  # Ensure agreement on secret

    # Deriving symmetric key using PBKDF2 (For Sender)
    salt = os.urandom(16)
    start_time = time.time()
    kdf_sender = PBKDF2HMAC(algorithm=hashes.SHA256(), length=32, salt=salt, iterations=100000)
    symmetric_key_sender = kdf_sender.derive(shared_secret_sender)
    key_derivation_time_sender = time.time() - start_time

    # AES Encryption
    padder = padding.PKCS7(128).padder()
    padded_data = padder.update(message) + padder.finalize()

    iv = os.urandom(16)  # Initialization Vector for AES
    start_time = time.time()
    cipher = Cipher(algorithms.AES(symmetric_key_sender), modes.CBC(iv))
    encryptor = cipher.encryptor()
    ciphertext = encryptor.update(padded_data) + encryptor.finalize()
    encryption_time = time.time() - start_time

    # Ciphertext size
    ciphertext_size = len(ciphertext)

    # AES Decryption (For Receiver)
    start_time = time.time()
    kdf_receiver = PBKDF2HMAC(algorithm=hashes.SHA256(), length=32, salt=salt, iterations=100000)
    symmetric_key_receiver = kdf_receiver.derive(shared_secret_receiver)
    key_derivation_time_receiver = time.time() - start_time

    # Decrypting the data with AES
    start_time = time.time()
    cipher = Cipher(algorithms.AES(symmetric_key_receiver), modes.CBC(iv))
    decryptor = cipher.decryptor()
    decrypted_padded_data = decryptor.update(ciphertext) + decryptor.finalize()
    decryption_time = time.time() - start_time

    # Unpadding the data
    unpadder = padding.PKCS7(128).unpadder()
    decrypted_data = unpadder.update(decrypted_padded_data) + unpadder.finalize()

    # Total key generation time
    total_key_gen_time = key_gen_time_sender + key_gen_time_receiver

    # Append results to DataFrame
    df = pd.concat([df, pd.DataFrame([{
        "Scheme": "ECC",
        "Message Size": size,
        "Key Generation Time": total_key_gen_time,
        "Encryption Time": encryption_time,
        "Decryption Time": decryption_time,
        "Ciphertext Size": ciphertext_size
    }])], ignore_index=True)

    # Print results
    print(f"‚úÖ ECC: {size} bytes - KeyGen: {total_key_gen_time:.6f}s, Enc: {encryption_time:.6f}s, Dec: {decryption_time:.6f}s, Ciphertext Size: {ciphertext_size} bytes")

# Save results back to the same file, keeping all data
df.to_csv(filename, index=False)

print(f"üìÑ ECC performance testing complete. Results saved in: {filename}")
```

```python
import time
import os
import oqs
import pandas as pd
from Crypto.Cipher import AES
from Crypto.Util.Padding import pad

# Define message sizes
message_sizes = [16, 32, 64, 128, 256, 512, 1024, 2048, 4096, 8192, 16384, 32768, 65536, 131072, 262144]

# Kyber variants to test
kyber_variants = ["Kyber512", "Kyber768", "Kyber1024"]

# File to store results
filename = "/content/performance_results.csv"

# Required columns
required_columns = ["Scheme", "Variant", "Message Size", "Key Generation Time",
                    "Encryption Time", "Decryption Time", "Ciphertext Size"]

# Load existing data safely
if os.path.exists(filename):
    df = pd.read_csv(filename)
    # Ensure all required columns exist
    for col in required_columns:
        if col not in df.columns:
            df[col] = None  # Add missing column with default values
else:
    df = pd.DataFrame(columns=required_columns)

# Ensure 'Scheme', 'Variant', and 'Message Size' exist before filtering
if not df.empty and all(col in df.columns for col in ["Scheme", "Variant", "Message Size"]):
    existing_kyber_results = set(zip(df["Scheme"], df["Variant"], df["Message Size"]))
else:
    existing_kyber_results = set()

for variant in kyber_variants:
    for size in message_sizes:
        if ("Kyber", variant, size) in existing_kyber_results:
            print(f"‚è≠ Skipping Kyber {variant} test for message size {size}, already in results.")
            continue

        # Generate random message
        message = os.urandom(size)

        # **Kyber Key Generation**
        kem = oqs.KeyEncapsulation(variant)
        start_time = time.time()
        public_key = kem.generate_keypair()
        private_key = kem.export_secret_key()
        keygen_time = time.time() - start_time

        # **Encapsulation (Encryption)**
        start_time = time.time()
        ciphertext, shared_secret = kem.encap_secret(public_key)  # Encrypt shared key

        # Encrypt message using AES-GCM with Kyber shared key
        aes_key = shared_secret[:32]  # Use first 32 bytes of shared secret
        iv = os.urandom(12)  # AES-GCM requires a 12-byte IV
        cipher = AES.new(aes_key, AES.MODE_GCM, iv)
        ciphertext_message, tag = cipher.encrypt_and_digest(message)
        encryption_time = time.time() - start_time

        # **Decapsulation (Decryption)**
        start_time = time.time()
        recovered_shared_secret = kem.decap_secret(ciphertext)
        decryption_time = time.time() - start_time

        # **Decrypt AES-GCM**
        aes_key = recovered_shared_secret[:32]
        cipher = AES.new(aes_key, AES.MODE_GCM, iv)
        decrypted_message = cipher.decrypt_and_verify(ciphertext_message, tag)

        # Verify correctness
        assert message == decrypted_message, "Decryption failed!"

        # Ciphertext size
        ciphertext_size = len(ciphertext) + len(ciphertext_message) + len(tag)

        # Append results to DataFrame
        df = pd.concat([df, pd.DataFrame([{
            "Scheme": "Kyber",
            "Variant": variant,
            "Message Size": size,
            "Key Generation Time": keygen_time,
            "Encryption Time": encryption_time,
            "Decryption Time": decryption_time,
            "Ciphertext Size": ciphertext_size
        }])], ignore_index=True)

        # Print progress
        print(f"‚úÖ Kyber-{variant}: {size} bytes - KeyGen: {keygen_time:.6f}s, Enc: {encryption_time:.6f}s, Dec: {decryption_time:.6f}s, Ciphertext Size: {ciphertext_size} bytes")

# Save results back to the file
df.to_csv(filename, index=False)

print(f"üìÑ Kyber performance testing complete. Results saved in: {filename}")
```

```python
import oqs
import time
import os
import pandas as pd
from Crypto.Cipher import AES
from Crypto.Util.Padding import pad, unpad

# Define message sizes
message_sizes = [16, 32, 64, 128, 256, 512, 1024, 2048, 4096, 8192, 16384, 32768, 65536, 131072, 262144]

# File to store results
results_file = "/content/performance_results.csv"

# Load existing results or create a new DataFrame
try:
    df = pd.read_csv(results_file)
except FileNotFoundError:
    df = pd.DataFrame(columns=["Variant", "Message Size", "Key Gen Time", "Encapsulation Time",
                               "Encryption Time", "Decapsulation Time", "Decryption Time", "Ciphertext Size"])

# Function to test a FrodoKEM variant
def test_frodokem(variant):
    kem = oqs.KeyEncapsulation(variant)

    for size in message_sizes:
        message = os.urandom(size)  # Generate a random message of the given size

        # Key Generation
        start_time = time.time()
        public_key = kem.generate_keypair()
        private_key = kem.export_secret_key()
        keygen_time = time.time() - start_time

        # Encapsulation (Key Exchange)
        start_time = time.time()
        ciphertext, shared_secret = kem.encap_secret(public_key)
        encapsulation_time = time.time() - start_time

        # Encrypt message using AES-GCM with the shared key
        aes_key = shared_secret[:32]  # First 32 bytes as AES key
        iv = os.urandom(12)  # Generate random IV for AES-GCM
        cipher = AES.new(aes_key, AES.MODE_GCM, iv)
        start_time = time.time()
        ciphertext_message, tag = cipher.encrypt_and_digest(pad(message, AES.block_size))
        encryption_time = time.time() - start_time

        # Decapsulation (Key Recovery)
        start_time = time.time()
        recovered_shared_secret = kem.decap_secret(ciphertext)
        decapsulation_time = time.time() - start_time

        # Decrypt AES-GCM message
        aes_key = recovered_shared_secret[:32]
        cipher = AES.new(aes_key, AES.MODE_GCM, iv)
        start_time = time.time()
        decrypted_message = unpad(cipher.decrypt_and_verify(ciphertext_message, tag), AES.block_size)
        decryption_time = time.time() - start_time

        # Verify decryption correctness
        assert decrypted_message == message, f"Decryption failed for {variant} with message size {size}!"

        # Calculate ciphertext size
        ciphertext_size = len(ciphertext) + len(ciphertext_message)

        # Store results
        df_new = pd.DataFrame([{
            "Variant": variant,
            "Message Size": size,
            "Key Gen Time": keygen_time,
            "Encapsulation Time": encapsulation_time,
            "Encryption Time": encryption_time,
            "Decapsulation Time": decapsulation_time,
            "Decryption Time": decryption_time,
            "Ciphertext Size": ciphertext_size
        }])

        global df
        df = pd.concat([df, df_new], ignore_index=True)

        # Print results
        print(f"‚úÖ {variant}: {size} bytes - KeyGen: {keygen_time:.6f}s, Encap: {encapsulation_time:.6f}s, "
              f"Enc: {encryption_time:.6f}s, Decap: {decapsulation_time:.6f}s, Dec: {decryption_time:.6f}s, "
              f"Ciphertext Size: {ciphertext_size} bytes")

# List of all FrodoKEM variants
frodokem_variants = [
    "FrodoKEM-640-AES", "FrodoKEM-640-SHAKE",
    "FrodoKEM-976-AES", "FrodoKEM-976-SHAKE",
    "FrodoKEM-1344-AES", "FrodoKEM-1344-SHAKE"
]

# Run tests for all FrodoKEM variants
for variant in frodokem_variants:
    test_frodokem(variant)

# Save results to file
df.to_csv(results_file, index=False)
print(f"‚úÖ FrodoKEM results saved to {results_file}")
```

```python
import time
import os
import oqs
import pandas as pd
from Crypto.Cipher import AES
from Crypto.Util.Padding import pad, unpad

# Define message sizes (same as Kyber)
message_sizes = [16, 32, 64, 128, 256, 512, 1024, 2048, 4096, 8192, 16384, 32768, 65536, 131072, 262144]

# NTRU variant to test
ntru_variant = "sntrup761"  # Supported in oqs-python

# File to store results
filename = "/content/performance_results.csv"

# Required columns
required_columns = ["Scheme", "Variant", "Message Size", "Key Generation Time",
                    "Encryption Time", "Decryption Time", "Ciphertext Size"]

# Load existing data if available
if os.path.exists(filename):
    df = pd.read_csv(filename)
    # Ensure all required columns exist
    for col in required_columns:
        if col not in df.columns:
            df[col] = None  # Add missing column with default values
else:
    df = pd.DataFrame(columns=required_columns)

# Convert existing results to a set for quick lookup
existing_results = set(zip(df["Scheme"], df["Variant"], df["Message Size"])) if "Variant" in df.columns else set()

# ‚úÖ Run tests for NTRU only
for size in message_sizes:
    if ("NTRU", ntru_variant, size) in existing_results:
        print(f"Skipping NTRU {ntru_variant} test for message size {size}, already in results.")
        continue  # Avoid duplicate testing

    # Generate random message
    message = os.urandom(size)

    # **Key Generation**
    kem = oqs.KeyEncapsulation(ntru_variant)
    start_time = time.time()
    public_key = kem.generate_keypair()
    private_key = kem.export_secret_key()
    keygen_time = time.time() - start_time

    # **Encapsulation (Encryption)**
    start_time = time.time()
    ciphertext, shared_secret = kem.encap_secret(public_key)

    # Encrypt message using AES-GCM with the shared secret
    aes_key = shared_secret[:32]  # First 32 bytes for AES
    iv = os.urandom(12)  # AES-GCM requires a 12-byte IV
    cipher = AES.new(aes_key, AES.MODE_GCM, iv)
    ciphertext_message, tag = cipher.encrypt_and_digest(pad(message, AES.block_size))
    encryption_time = time.time() - start_time

    # **Decapsulation (Decryption)**
    start_time = time.time()
    recovered_shared_secret = kem.decap_secret(ciphertext)
    decryption_time = time.time() - start_time

    # **Decrypt AES-GCM**
    aes_key = recovered_shared_secret[:32]
    cipher = AES.new(aes_key, AES.MODE_GCM, iv)
    decrypted_message = unpad(cipher.decrypt_and_verify(ciphertext_message, tag), AES.block_size)

    # Ciphertext size
    ciphertext_size = len(ciphertext) + len(ciphertext_message)

    # ‚úÖ Print metrics
    print(f"\nNTRU Variant: {ntru_variant}, Message Size: {size} bytes")
    print(f"  Key Generation Time: {keygen_time:.6f} seconds")
    print(f"  Encryption Time: {encryption_time:.6f} seconds")
    print(f"  Decryption Time: {decryption_time:.6f} seconds")
    print(f"  Ciphertext Size: {ciphertext_size} bytes")

    # ‚úÖ Save results
    df = pd.concat([df, pd.DataFrame([{
        "Scheme": "NTRU",
        "Variant": ntru_variant,
        "Message Size": size,
        "Key Generation Time": keygen_time,
        "Encryption Time": encryption_time,
        "Decryption Time": decryption_time,
        "Ciphertext Size": ciphertext_size
    }])], ignore_index=True)

    # ‚úÖ Ensure correctness
    assert message == decrypted_message, "Decryption failed!"

# Save results
df.to_csv(filename, index=False)
print(f"\nNTRU performance testing complete. Results saved in: {filename}")
```
